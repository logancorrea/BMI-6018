```python
# -*- coding: utf-8 -*-

#%% Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#%% Importing Data
flights_data = pd.read_csv(r'C:\Users\logan\Documents\Intro to Programming\M09\flights.csv')
flights_data.head(10)
weather_data_pd = pd.read_csv('weather.csv')
weather_data_np = weather_data_pd.to_numpy()
fd = flights_data
fd






```

```python
#Question 1 How many flights were there from JFK to SLC? Int
jfk_slc = fd.loc[(fd['origin'] == "JFK") & (fd['dest'] == "SLC")]
q_1 = len(jfk_slc)
q_1
```

```python
#Question 2 How many airlines fly to SLC? Should be int
slc = fd.loc[fd['dest'] == "SLC"]
airline = slc.carrier.unique()
q_2 = len(airline)
q_2
```

```python
#Question 3 What is the average arrival delay for flights to RDU? float
rdu = fd.loc[fd['dest'] == "RDU"]
q_3 = rdu['arr_delay']
q_3.mean()
```

```python
#Question 4 What proportion of flights to SEA come from the two NYC airports (LGA and JFK)?  float
sea = fd.loc[fd['dest'] == "SEA"]
lga = sea.loc[fd['origin'] == "LGA"]
jfk = sea.loc[fd['origin'] == "JFK"]
sea_sum = len(sea)
lga_sum = len(lga)
jfk_sum = len(jfk)
q_4 = (lga_sum + jfk_sum)/sea_sum
q_4
```

```python
#Question 5 Which date has the largest average depature delay? Pd slice with date and float
#please make date a column. Preferred format is 2013/1/1 (y/m/d)
fd['date'] = fd.year.astype(str) + "/" + fd.month.astype(str) + "/" + fd.day.astype(str)
delay = fd[['date', 'dep_delay']]
average = delay.groupby(delay.date).mean()
max_index = average['dep_delay'].idxmax()
q_5 = average.loc[max_index]
q_5
```

```python
#Question 6 Which date has the largest average arrival delay? pd slice with date and float
arr_delay = fd[['date', 'arr_delay']]
arr_average = arr_delay.groupby(arr_delay.date).mean()
arr_max_index = arr_average['arr_delay'].idxmax()
q_6 = arr_average.loc[arr_max_index]
q_6
```

```python
#Question 7 Which flight departing LGA or JFK in 2013 flew the fastest? pd slice with tailnumber and speed
#speed = distance/airtime
fd['speed'] = fd.distance/fd.arr_time
lga_jfk = fd.loc[(fd['origin'] == "LGA") | (fd['origin'] == "JFK")]
fastest = lga_jfk[['tailnum','speed']]
fastest_index = fastest['speed'].idxmax()
q_7 = fastest.loc[fastest_index]
q_7
```
