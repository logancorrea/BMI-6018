```python
import pandas as pd
df = pd.read_csv("breast-cancer.data", header=None)
df.columns = ['Class', 'age', 'menopause', 'tumor-size', 'inv-nodes', 'node-caps', 'deg-malig', 'breast', 'breast-quad', 'irradiat']
df
```

```python
#Melt
melt_columns = ['age', 'menopause', 'tumor-size']
df_melted = df.melt(id_vars=melt_columns, var_name='variable', value_name='value')
df_melted
```

```python
#Pivot
df1 = df.drop_duplicates(subset=['age'])
df_pivot = df1.pivot(index='age', columns='breast-quad', values='irradiat')
df_pivot
```

```python
#Aggregate
df.agg({ 'age' : "count", 'deg-malig' : "max", 'tumor-size' : "min"})
```

```python
#Iteration
for index, row in df.iterrows():
    print(index, row['age'])
```

```python
#Groupby
grouped = df.groupby('Class')
max = grouped['tumor-size'].max()
max
```
