BioPython is an open-source python library that contains tools for biological computation. The source code is currently released under the Biopython License. Advantages of BioPython include extensive file format functionality and online database access to provide easy manipulation of biological sequence data. One disadvantage of BioPython is that it's performance may not be optimized for specialized tasks that might users to use higher-level coding languages.


```python
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord

# Read the FASTA file
fasta_file_path = r"C:\Users\logan\Documents\Intro to Programming\M11\fake_fasta.fasta"
records = SeqIO.to_dict(SeqIO.parse(fasta_file_path, "fasta"))

# Translate each sequence and create new SeqRecord objects
translated_records = []
for record_id, record in records.items():
    translated_sequence = str(record.seq.translate())
    translated_record = SeqRecord(Seq(translated_sequence), id=record_id, description=record.description)
    translated_records.append(translated_record)

# Print translated sequences
for translated_record in translated_records:
    print(f"ID: {translated_record.id}")
    print(f"Description: {translated_record.description}")
    print(f"Translated Sequence: {translated_record.seq}")
    print("\n")
```
